```python
a_2d = np.array([2, 5])
b_2d = np.array([1, 2])
sum_2d = a_2d + b_2d

plt.figure(figsize = (8, 6))
plt.quiver(0, 0, a_2d[0], a_2d[1], angles = 'xy', scale_units='xy', scale =1, color='red', width = 0.005)
plt.quiver(0, 0, b_2d[0], b_2d[1], angles='xy', scale_units='xy', scale=1, color='blue', width=0.005)
plt.quiver(0, 0, sum_2d[0], sum_2d[1], angles='xy', scale_units='xy', scale=1, color='green', width=0.005)
plt.quiver(a_2d[0], a_2d[1], b_2d[0], b_2d[1], angles='xy', scale_units='xy', scale=1, color='blue', width=0.003, alpha=0.5)
plt.xlim(-1, 4)
plt.ylim(-1, 8)
plt.grid(True)
plt.title("Vector Addition: Parallelogram Rule")
plt.legend(['Vector a', 'Vector b', 'a + b'])

plt.show()
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Properties of vector addition
c = np.array([1, -1, 2])

# Commutative property: a + b = b + a
print(f"a + b = {a + b}")
print(f"b + a = {b + a}")
print(f"Commutative: {np.array_equal(a + b, b + a)}")

# Associative property: (a + b) + c = a + (b + c)
left = (a + b) + c
right = a + (b + c)
print(f"(a + b) + c = {left}")
print(f"a + (b + c) = {right}")
print(f"Associative: {np.array_equal(left, right)}")
```

**The Physicist's Perspective of Vectors**

- Vectors have both magnitude and directions in space

```python
import matplotlib.pyplot as plt
import numpy as np
vector_a = np.array([3, 4])
magnitude = np.linalg.norm(vector_a)

print(f"Vector : {vector_a}")
print(f"Magnitude : {magnitude}")

unit_vector = vector_a/magnitude
print(f"Unit Vector: {unit_vector}")
```

```python
plt.figure(figsize=(8, 6))
plt.quiver(0, 0, vector_a[0], vector_a[1], angles='xy', scale_units='xy', scale=1, color='red', width =0.005)
plt.quiver(0, 0, unit_vector[0], unit_vector[1], angles='xy', scale_units='xy', scale=1, color='blue', width =0.005)
plt.xlim(-1, 5)
plt.xlabel("Direction in x-axis")
plt.ylim(-1, 5)
plt.grid(True)
plt.title("Vector: Magnitude and Direction")
plt.legend(['Vector A', 'Unit Vector of A'])
plt.show()
```

**The Mathematician's Perspective**
- Abstract Objects
 - Vectors are any objects that support addition and scalar multiplication.

```python
poly1 = np.array([1, 2, 3])
poly2 = np.array([2, -1, 1])

print(f"Sum : {poly1+poly2}")

# Example 2: Functions as vectors (function space)
# This is more abstract, but functions can be added and scaled
def func1(x):
    return x**2

def func2(x):
    return 2*x

# Function addition: (f + g)(x) = f(x) + g(x)
def func_sum(x):
    return func1(x) + func2(x)

x_vals = np.linspace(0, 3, 100) # Generate 100 points beween 0 and 3

plt.figure(figsize=(10, 6))
plt.plot(x_vals, func1(x_vals), label='f(x) = x²') 
plt.plot(x_vals, func2(x_vals), label='f(x) = 2x') 
plt.plot(x_vals, func_sum(x_vals), label='f(x) = x² + 2x') 
plt.legend(['x²', '2x', 'x² + 2x'])
plt.grid()
plt.title('Functions as Vectors')

plt.show()
```

```python
# 3D geometric vectors
position = np.array([10, 5, 2])  # Position in 3D space
velocity = np.array([2, -1, 3])   # Velocity vector
force = np.array([0, 0, -9.8])    # Gravitational force

print("Position vector:", position)
print("Velocity vector:", velocity)
print("Force vector:", force)

# Calculate new position after time t=1
new_position = position + velocity
print("New position after 1 unit time:", new_position)
```

```python
# Audio signals as vectors
sample_rate = 44100
duration = 0.1  # 0.1 seconds
t = np.linspace(0, duration, int(sample_rate * duration))

# Two sine waves (audio signals)
signal1 = 0.5 * np.sin(2 * np.pi * 440 * t)  # A note (440 Hz)
signal2 = 0.3 * np.sin(2 * np.pi * 880 * t)  # A note one octave higher

# Vector addition of audio signals
mixed_signal = signal1 + signal2

# Scalar multiplication (volume control)
quiet_signal = 0.5 * mixed_signal

plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.plot(t[:1000], signal1[:1000])
plt.title("Signal 1 (440 Hz)")
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(t[:1000], signal2[:1000])
plt.title("Signal 2 (880 Hz)")
plt.grid(True)

plt.subplot(2, 2, 3)
plt.plot(t[:1000], mixed_signal[:1000])
plt.title("Mixed Signal (Addition)")
plt.grid(True)

plt.subplot(2, 2, 4)
plt.plot(t[:1000], quiet_signal[:1000])
plt.title("Quiet Signal (Scalar Multiplication)")
plt.grid(True)

plt.tight_layout()
plt.show()

print(f"Signal 1 shape: {signal1.shape}")
print(f"Signal 2 shape: {signal2.shape}")
print(f"Mixed signal shape: {mixed_signal.shape}")
```

```python
# 1. Word frequencies in documents (Real-world)
document1 = np.array([5, 3, 0, 2, 1])  # word counts: [the, cat, dog, sat, ran]
document2 = np.array([2, 0, 4, 1, 3])  # word counts for same vocabulary

# Document similarity (dot product)
similarity = np.dot(document1, document2)
print("Document vectors:")
print("Doc 1:", document1)
print("Doc 2:", document2)
print("Similarity score:", similarity)
```

```python
# 2. Stock prices over time (Real-world)
stock_prices = np.array([100, 102, 98, 105, 103, 107])
print("\nStock price vector:", stock_prices)

# Price changes (difference vector)
price_changes = np.diff(stock_prices)
print("Price changes:", price_changes)
```

```python
# 3. Probability distributions (Abstract)
prob_dist = np.array([0.1, 0.2, 0.3, 0.25, 0.15])
print("\nProbability distribution:", prob_dist)
print("Sum (should be 1):", np.sum(prob_dist))

# Scaling maintains probability properties
scaled_prob = 2 * prob_dist
print("Scaled (no longer probability):", scaled_prob)
print("Sum:", np.sum(scaled_prob))
```

```python
# Project Setup 
# import neccesary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

print("imported successes ")
```

```python
filepath = Path('eletronic_sales.xlsx')
sales = pd.read_excel(filepath)
if not sales.empty: print("Successfull")
```

```python
sales.head()
```

```python
# add revenue column
# To create a new column, you start with creating the new columns as a variable name

# units * price = Revenue
sales['Revenue'] = sales['Units'] * sales['Price']
sales
```

```python
sales["Revenue(N)"] = round(sales['Revenue']*1000, 2)
sales
```

```python
sales['Date'] = pd.to_datetime(sales['Date'])
sales.dtypes
```

```python
# lets just plot  this, its not important though.
sales.dtypes.value_counts().plot(kind='bar', xlabel='D-Types', ylabel='Number')
```

```python
sales['Year'] = sales['Date'].dt.year
sales['Month'] = sales['Date'].dt.month_name()
sales['Day'] = sales['Date'].dt.day_name()
sales.head()
```

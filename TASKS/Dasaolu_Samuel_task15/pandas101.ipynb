```python
# --- PANDAS 101 AS TASK ---

import pandas as pd
import matplotlib.pyplot as plt

sales = pd.read_csv("sales_data.xls")
sales.isna().sum()
```

```python
sales.loc[[0, 1,4, 5]]
```

```python
sales.iloc[[0, 1,4, 5]] # note that iloc resets index by default
```

```python
sales.loc[[0,2,4,6],['customer_id','device_id','item_id']]
```

```python
# sales_data.iloc[[0,2,4,6],['customer_id','device_id','item_id']]
# using iloc with non-number indexes, throw error
```

```python
sales.iloc[[0,2,4,6],[0,1,2,3]]
```

```python
sales.iloc[::] #returns everything
sales.iloc[:] # also returns everything
# sales_data.iloc[0:-1] #also returns everything
```

```python
sales.iloc[-1:] # i loc returns only last element, loc returns all elements
sales.loc[-1:]
```

```python
# Lets use loc and iloc to slice columns...
sales.loc[0:4,'customer_id':'item_id'] # Pick from row 0 to 4, and column 1 to 3...
```

```python
# Using iloc for integer or index numbers
sales.iloc[0:4,0:4]
```

```python
# lets work with index...using loc and iloc...
sales.set_index("location", inplace = True), #inplace brings the index column to the beginning
sales
```

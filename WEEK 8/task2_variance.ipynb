# CLASS TASK 2: VARIANCE


**Given**:

```
[4, 8, 6, 5, 3]
```

Number of items:  
\( N = 5 \)

**Mean:**
$ \overline{x} = \frac{4 + 8+ 6 + 5 + 3}{5} = 5.2 $

$ \therefore \overline{x} = \frac{26}{5} = 5.2 $

---

### Workings

| $x$ | $x - \overline{x} $ | $(x - \overline{x})^2 $ |
|-----|---------------------|-------------------------|
| $4$ | $4 - 5.2 = -1.2$    | $(-1.2)^2 = 1.44$       |
| $8$ | $8 - 5.2 =  2.8$    | $(2.8)^2 = 7.84 $       |
| $6$ | $6 - 5.2 =  0.8$    | $(0.8)^2 = 0.64 $       |
| $5$ | $5 - 5.2 = -0.2$    | $(-0.2)^2 = 0.04$       |
| $3$ | $3 - 5.2 = -2.2$    | $(-2.2)^2 = 4.84$       |

---

**Sum of squared deviations:**

$ \sum (x - \bar{x})^2 = 1.44 + 7.84 + 0.64 + 0.04 + 4.84 = 14.8 $

---

### Variance

- **Population variance:**
$ \sigma^2 = \frac{14.8}{5} = 2.96 $


# $\therefore$
### Population variance = **2.96**
